```python
import pandas as pd
```

```python
data = pd.read_csv("/content/drive/MyDrive/Data Science/area.csv")
```

```python
data.head()
```

Splitting data into independent and dependent variable

```python
x = data[['area']]
```

```python
x
```

```python
y = data.price
```

```python
y
```

Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(x,y, color = "red", s = 100, ec = ["yellow","blue"])
plt.show()
```

```python
pip install scikit-learn
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[5000]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.coef_ * 5000 + model.intercept_
```

```python
pred = model.predict(x)
```

```python
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(x,y, color = "red", s = 100, ec = ["yellow","blue"])
plt.plot(x, pred, color = "green")
plt.show()
```
