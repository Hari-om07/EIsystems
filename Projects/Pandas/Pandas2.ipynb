Handling the missing values using filling rows and columns

```python
data2.fillna("hello")
```

```python
data2.fillna(data2["price"].mean()) #puts the mean of price in missing space
```

```python
data2.fillna({
    "clarity":"bye",
    "color":"world"
})
```

```python
data2.fillna(method="bfill") #fills the data of next row in place of missing data or backwards the values in place of missing data
```

```python
data2.fillna(method = "ffill") #fills the data of previous row in place of missing data or forwards the values in place of missing data
```

```python
data2.fillna(method = "ffill", limit = 2) #sets the limit for forwarding the data
```

Merging two or more dataframe

```python
data1 = pd.DataFrame({
    "name":["a", "b", "c", "d", "e"],
    "age":[11, 22, 33, 44, 55],
    "sub":["s1", "s2", "s3", "s4", "s5"]
})
data2 = pd.DataFrame({
    "name":["a", "bb", "c", "d", "e"],
    "age":[11, 22, 33, 444, 55],
    "sub":["s1", "s2", "s33", "s4", "s5"]
})
```

```python
data1
```

```python
data2
```

```python
DATA = pd.concat([data1,data2], ignore_index = True, axis = 0)
```

```python
DATA
```

```python
DATA = pd.merge(data1,data2) #shows the data common in both dataset
```

```python
DATA
```

```python
DATA = pd.merge(data1,data2, on = "name" ) #shows the data common in both dataset considering only "name" column
```

```python
DATA
```

```python
DATA = pd.merge(data1,data2, on = "name", how = "outer") #shows the common data in both dataset and uncommon as well considering "name" column
```

```python
DATA
```

```python
DATA = pd.merge(data1,data2, on = "name", how = "left") #shows the data of dataset1 and common in both considering "name" column
```

```python
DATA
```

```python
DATA = pd.merge(data1,data2, on = "name", how = "right") #shows the data of dataset2 and common in both considering "name" column
```

```python
DATA
```
