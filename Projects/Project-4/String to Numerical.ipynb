```python
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/Data Science/Bmi_male_female.csv')
```

```python
data.head()
```

```python
x = data[["Height", "Weight"]]
y = data.Gender
```

```python
#Converting into numerical using one-hot encoding
Y = pd.get_dummies(y)
```

```python
Y
```

```python
from sklearn.tree import DecisionTreeClassifier
import warnings
warnings.filterwarnings('ignore')
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x,Y)
```

```python
prediction = model.predict([[190,60]])
```

```python
if prediction[0][0]==0:
  print("Female")
else:
  print("Male")
```
